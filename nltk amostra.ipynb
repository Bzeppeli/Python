```python
sent = 'Mr. Green killed colonel Mustard in the study woth the candlestick. Mr. Green is not a very nice follow'
```

```python
#divide em sentenças separadas por pontos
phrase = nltk.tokenize.sent_tokenize(sent)
```

```python
phrase
```

```python
tokens = nltk.word_tokenize(sent)
```

```python
tokens
```

```python
#indentifica as classes gramaticais
classes = nltk.pos_tag(tokens)
```

```python
classes
```

```python
#e o tipo de entidade que elas pertencem, se é pessoa, orgão, etc...
entities = nltk.chunk.ne_chunk(classes)
```

```python
print(entities)
```
